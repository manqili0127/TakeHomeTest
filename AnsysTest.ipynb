A. Design a dummy pandas data frame with 6 columns, with header names X. Y, Z.meltpool(mp) length, mp width and mp depth with random data of column length 10, findmaximum, minimum and calculate average mp length, mp depth and mp width, and plotcolumn X vs meltpool depth, and X vs meltpool width.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a pandas DataFrame with random data
np.random.seed(40)  # Setting seed for reproducibility
data = {
    'X': np.random.rand(10),
    'Y': np.random.rand(10),
    'Z': np.random.rand(10),
    'meltpool_length': np.random.rand(10),
    'meltpool_width': np.random.rand(10),
    'meltpool_depth': np.random.rand(10),
}

df = pd.DataFrame(data)
df
```

```python
# Find maximum and minimum for each column
max_values = df.max()
min_values = df.min()

# Create a new DataFrame for maximum and minimum values
max_min_df = pd.DataFrame({
    'max': max_values,
    'min': min_values
})

# exchange row and column for better visualization
max_min_df = max_min_df.transpose()
max_min_df
```

```python
# Calculate average values
avg_mp_length = df['meltpool_length'].mean()
avg_mp_width = df['meltpool_width'].mean()
avg_mp_depth = df['meltpool_depth'].mean()

# Create a new DataFrame for averages
avg_df = pd.DataFrame({
    'avg_mp_length': [avg_mp_length],
    'avg_mp_width': [avg_mp_width],
    'avg_mp_depth': [avg_mp_depth],
})

avg_df
```

```python
# Plotting
plt.figure(figsize=(8, 4))

# Plot X vs meltpool depth
plt.subplot(1, 2, 1)
plt.scatter(df['X'], df['meltpool_depth'])
plt.title('X vs Meltpool Depth')
plt.xlabel('X')
plt.ylabel('Meltpool Depth')

# Plot X vs meltpool width
plt.subplot(1, 2, 2)
plt.scatter(df['X'], df['meltpool_width'])
plt.title('X vs Meltpool Width')
plt.xlabel('X')
plt.ylabel('Meltpool Width')

plt.tight_layout()
plt.show()
```

B. Design a parametric design table with three sets of parameters, Laser Power (100, 200, 300 W), Laser Speed (800, 1000, 1500, 2000 mm/s) and layer thickness (30e-6, 40e-6microns), Print total combinations of these parameters and filter the parameters rows withpowers (200, 300 W) and speed (1000, 1500, 2000 mm/s ) from the total combinationsand finally write a csv with filtered parameters.

```python
import pandas as pd
from itertools import product

# Define parameter values
laser_power_values = [100, 200, 300]
laser_speed_values = [800, 1000, 1500, 2000]
layer_thickness_values = [30e-6, 40e-6]

# Generate all combinations of parameters
all_combinations = list(product(laser_power_values, laser_speed_values, layer_thickness_values))

# Create a DataFrame with all combinations
column_names = ['Laser Power (W)', 'Laser Speed (mm/s)', 'Layer Thickness (microns)']
parametric_table = pd.DataFrame(all_combinations, columns=column_names)
parametric_table
```

```python
# Filter rows based on specified criteria
filtered_table = parametric_table[
    (parametric_table['Laser Power (W)'].isin([200, 300])) &
    (parametric_table['Laser Speed (mm/s)'].isin([1000, 1500, 2000]))
]

# Display the filtered DataFrame
filtered_table
```

```python
# Write the filtered parameters to a CSV file
filtered_table.to_csv('filtered_parameters.csv', index=False)
```
